```python
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
scores      = [40, 50, 55, 65, 70, 80, 85, 95]
```

```python
#reshape the data
X = np.array(study_hours).reshape(-1,1)
y = np.array(scores)

#build the model
model = LinearRegression()
#train the model
model.fit(X,y)
#Predict
pred = model.predict([[10]])


print(f"IF 10 hour study the score gonna be : {pred[0]:.1f}")
print(f"Slope(m)(weight) : {model.coef_[0]:.2f}")
print(f"Intercept(Bias) : {model.intercept_:.2f}")
```
